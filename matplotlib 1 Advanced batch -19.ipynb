```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[7,8,9]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[7,8,9]

# adding the labels

plt.xlabel('sales')
plt.ylabel('revenue')

plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[7,8,9]
z=[4,5,6]

# adding the labels

plt.xlabel('sales')
plt.ylabel('revenue')
plt.title('Sales performance')
plt.plot(x,y)

# adding a new plot layer x and z
plt.plot(x,z)
plt.show()
```

```python
x=[1,2,3]
y=[7,8,9]
z=[4,5,6]

# adding the labels

plt.xlabel('sales')
plt.ylabel('revenue')
plt.title('Sales performance')
plt.plot(x,y)

# adding a new plot layer x and z
plt.plot(x,z)
plt.legend(['this is x and y','this is x and z'])
plt.show()
```

```python
# sample data set
x=[1,2,3]
y=[7,8,9]
z=[4,5,6]

# adding the labels

plt.xlabel('sales')  # adding x lable
plt.ylabel('revenue')  # adding y label
plt.title('Sales performance')  # adding title of the graph
plt.plot(x,y)    # doing the plotting of x and y

# adding a new plot layer x and z
plt.plot(x,z)    # doing the plotting of x and z
plt.legend(['this is x and y','this is x and z'])  # adding the legend to show the difference between layers
plt.show()    # showing the graph view
```
